```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.colors as colors

%matplotlib inline

from matplotlib import rc

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)

from classical_networks import *

fs = 20
```

## Distribution of Coefficients $U_{jk}$ and $V_{jl}^{(k)}$

Refer to Figures 7 and 8 in the appendices of https://arxiv.org/pdf/1810.04178.pdf for comparison with a separate implementation of this model.

```python
N = 40

reps = 120

allEs_h1 = np.array([])
allUs_h1 = np.array([])
allVs_h1 = np.array([])

allEs_h6 = np.array([])
allUs_h6 = np.array([])
allVs_h6 = np.array([])

for i in range(reps):
    if i % (reps // 10) == 0:
        print(i)

    (e,u,v) = generateRealization(N,1)
    v = v.flatten()
    v = v[np.abs(v) > 0]
    allEs_h1 = np.append(allEs_h1,e)
    allUs_h1 = np.append(allUs_h1, u.flatten())
    allVs_h1 = np.append(allVs_h1, v.flatten())

    (e,u,v) = generateRealization(N,6)
    v = v.flatten()
    v = v[np.abs(v) > 0]
    allEs_h6 = np.append(allEs_h6,e)
    allUs_h6 = np.append(allUs_h6, u.flatten())
    allVs_h6 = np.append(allVs_h6, v.flatten())
```

```python
plt.rc('xtick',labelsize=fs)
plt.rc('ytick',labelsize=fs)

fig,ax = plt.subplots(1,3,figsize=(16,4.5))

lw = 2.5
b = 100

# h = 1
ax[0].hist(allEs_h1,bins=b,density=True,histtype='step',lw=lw)
ax[1].hist(allUs_h1,bins=b,density=True,histtype='step',lw=lw,label=r'$1$')
ax[2].hist(allVs_h1,bins=b,density=True,histtype='step',lw=lw)

# h = 6
ax[0].hist(allEs_h6,bins=b,density=True,histtype='step',lw=lw)
ax[1].hist(allUs_h6,bins=b,density=True,histtype='step',lw=lw,label=r'$6$')
ax[2].hist(allVs_h6,bins=b,density=True,histtype='step',lw=lw)

ax[1].set_yscale('log')
ax[2].set_yscale('log')

ax[0].set_xlabel(r'$\varepsilon_j$',fontsize=fs)
ax[1].set_xlabel(r'$U_{jl}$',fontsize=fs)
ax[2].set_xlabel(r'$V_{jl}^{(k)}$',fontsize=fs)

leg = ax[1].legend(fontsize=fs-2,title=r'$h$')
leg.get_title().set_fontsize(fs-2)

fig.text(0.49, 0.92, r'$N=40$',fontsize=fs)

fig.subplots_adjust(wspace=.3)
```

```python
N = 10
h = 6

reps = 200

allEs_L10 = np.array([])
allUs_L10 = np.array([])
allVs_L10 = np.array([])

allEs_L20 = np.array([])
allUs_L20 = np.array([])
allVs_L20 = np.array([])

allEs_L30 = np.array([])
allUs_L30 = np.array([])
allVs_L30 = np.array([])

allEs_L40 = np.array([])
allUs_L40 = np.array([])
allVs_L40 = np.array([])

for i in range(reps):
    if i % (reps // 10) == 0:
        print(i)

    (e,u,v) = generateRealization(10,h)
    v = v.flatten()
    v = v[np.abs(v) > 0]
    allEs_L10 = np.append(allEs_L10,e)
    allUs_L10 = np.append(allUs_L10, u.flatten())
    allVs_L10 = np.append(allVs_L10, v.flatten())

    (e,u,v) = generateRealization(20,h)
    v = v.flatten()
    v = v[np.abs(v) > 0]
    allEs_L20 = np.append(allEs_L20,e)
    allUs_L20 = np.append(allUs_L20, u.flatten())
    allVs_L20 = np.append(allVs_L20, v.flatten())

    (e,u,v) = generateRealization(30,h)
    v = v.flatten()
    v = v[np.abs(v) > 0]
    allEs_L30 = np.append(allEs_L30,e)
    allUs_L30 = np.append(allUs_L30, u.flatten())
    allVs_L30 = np.append(allVs_L30, v.flatten())
    
    (e,u,v) = generateRealization(40,h)
    v = v.flatten()
    v = v[np.abs(v) > 0]
    allEs_L40 = np.append(allEs_L40,e)
    allUs_L40 = np.append(allUs_L40, u.flatten())
    allVs_L40 = np.append(allVs_L40, v.flatten())
    
```

```python
plt.rc('xtick',labelsize=20)
plt.rc('ytick',labelsize=20)

fig,ax = plt.subplots(figsize=(9,4.5))

ax.hist(np.log10(allUs_L10),bins=b,density=True,histtype='step',lw=2.5, label=r'$10$')
ax.hist(np.log10(allUs_L20),bins=b,density=True,histtype='step',lw=2.5, label=r'$20$')
ax.hist(np.log10(allUs_L30),bins=b,density=True,histtype='step',lw=2.5, label=r'$30$')
ax.hist(np.log10(allUs_L40),bins=b,density=True,histtype='step',lw=2.5, label=r'$40$')
ax.set_yscale('log')
leg = ax.legend(title=r'$L$',fontsize=fs-2)
leg.get_title().set_fontsize(fs-2)

ax.set_xlabel(r'$x = \log_{10}\left(U_{jl}\right)$',fontsize=fs)
ax.set_ylabel(r'$P(x)$',fontsize=fs)
```

```python
plt.rc('xtick',labelsize=20)
plt.rc('ytick',labelsize=20)

fig,ax = plt.subplots(figsize=(9,4.5))

ax.hist(np.log10(np.abs(allVs_L10)),bins=b,density=True,histtype='step',lw=3,label=r'$10$')
ax.hist(np.log10(np.abs(allVs_L20)),bins=b,density=True,histtype='step',lw=3,label=r'$20$')
ax.hist(np.log10(np.abs(allVs_L30)),bins=b,density=True,histtype='step',lw=3,label=r'$30$')
ax.hist(np.log10(np.abs(allVs_L40)),bins=b,density=True,histtype='step',lw=3,label=r'$40$')

ax.set_yscale('log')
leg = ax.legend(title=r'$L$',fontsize=fs-2)
leg.get_title().set_fontsize(fs-2)

ax.set_xlabel(r'$x = \log_{10}\left(\left\vert V_{jl}^{(k)}\right\vert\right)$',fontsize=fs)
ax.set_ylabel(r'$P(x)$',fontsize=fs)
```

```python
plt.rc('xtick',labelsize=20)
plt.rc('ytick',labelsize=20)

fig,ax = plt.subplots(figsize=(9,4.5))

lw = 2.5

(y,x) = np.histogram(np.log10(np.abs(allVs_L10)),bins=b,density=True)
x = x[1:]
y *= (10**2)
plt.plot(x,y,lw=lw,label=r'$10$')

(y,x) = np.histogram(np.log10(np.abs(allVs_L20)),bins=b,density=True)
x = x[1:]
y *= (20**2)
plt.plot(x,y,lw=lw,label=r'$20$')

(y,x) = np.histogram(np.log10(np.abs(allVs_L30)),bins=b,density=True)
x = x[1:]
y *= (30**2)
plt.plot(x,y,lw=lw,label=r'$30$')

(y,x) = np.histogram(np.log10(np.abs(allVs_L40)),bins=b,density=True)
x = x[1:]
y *= (40**2)
plt.plot(x,y,lw=lw,label=r'$40$')

leg = plt.legend(title=r'$L$',fontsize=fs)
leg.get_title().set_fontsize(fs-2)

plt.xlabel(r'$x = \log_{10}\left(\left\vert V_{jl}^{(k)} \right\vert \right)$',fontsize=fs)
plt.ylabel(r'$P(x) \cdot L^2$',fontsize=fs)
```
